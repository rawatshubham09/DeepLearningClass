```python
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

import tensorflow
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense, Dropout
from tensorflow.keras.optimizers import Adam

from sklearn.datasets import make_moons
from sklearn.model_selection import train_test_split

from mlxtend.plotting import plot_decision_regions
```

```python
# Creating datasets

X,y = make_moons(100, random_state=2, noise=0.25)
```

```python
X[:4]
```

```python
plt.rcParams["figure.figsize"] = (10,8)

plt.scatter(X[:,0], X[:,1], c=y)
plt.show()
```

```python
X_train,X_test,y_train,y_test = train_test_split(X,y,test_size=0.2, random_state=123)
```

## Without Using Regularization

```python
model = Sequential()

model.add(Dense(128, activation="relu", input_dim=2))
model.add(Dense(64, activation="relu"))
model.add(Dense(1, activation="sigmoid"))

model.summary()
```

```python
adam = Adam(0.001)

model.compile(loss="binary_crossentropy", optimizer=adam, metrics=["accuracy"])
```

```python
history = model.fit(X_train,y_train, validation_data=(X_test,y_test),epochs=2000)
```

```python
plot_decision_regions(X, y.astype('int'), clf=model, legend=2)
plt.xlim(-2,3)
plt.ylim(-1.5,2)
plt.show()
```

```python
# Plotting val-loss and trainning loss
plt.plot(history.history['loss'])
plt.plot(history.history['val_loss'])
```

## With Regularization

```python
model1 = Sequential()

model1.add(Dense(128, activation="relu", input_dim=2, kernel_regularizer=tensorflow.keras.regularizers.l2(0.01)))
model1.add(Dense(64, activation="relu", kernel_regularizer=tensorflow.keras.regularizers.l2(0.01)))
model1.add(Dense(1, activation="sigmoid"))

model1.summary()
```

```python
model1.compile(loss="binary_crossentropy", optimizer=adam, metrics=["accuracy"])
```

```python
history = model1.fit(X_train,y_train, validation_data=(X_test,y_test),epochs=2000)
```

```python
plot_decision_regions(X, y.astype('int'), clf=model1, legend=2)
plt.xlim(-2,3)
plt.ylim(-1.5,2)
plt.show()
```

```python
# Plotting val-loss and trainning loss
plt.plot(history.history['loss'])
plt.plot(history.history['val_loss'])
```

```python
# Weights of previous and new layers
model1_weight_layer1 = model.get_weights()[0].reshape(256)
model2_weight_layer1 = model1.get_weights()[0].reshape(256)
```
