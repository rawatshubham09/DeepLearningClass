```python
import numpy as np
import matplotlib.pyplot as plt

import tensorflow
from sklearn.datasets import make_circles
from tensorflow import keras
from keras.layers import Dense, BatchNormalization
from keras.models import Sequential

%matplotlib inline
```

```python
plt.rcParams["figure.figsize"]=(10,8)
```

```python
X,y = make_circles(n_samples=200, noise=0.1)
```

```python
plt.scatter(X[:,0],X[:,1],c=y)
```

## Without batch normalization

```python
model = Sequential()

model.add(Dense(4, input_dim=2, activation='relu'))
model.add(Dense(2,activation='relu'))
model.add(Dense(1,activation='sigmoid'))

model.summary()
```

```python
model.compile(loss='binary_crossentropy',optimizer='adam',metrics=['accuracy'])
```

```python
history = model.fit(X,y,epochs=200,validation_split=0.2)
```

```python
# plotting loss
plt.plot(history.history["val_loss"],color="red",label="Val_loss")
plt.plot(history.history["loss"],color="green",label="loss")
plt.legend()
```

```python
model.get_weights()
```

# With Batch Normalization

```python
model1 = Sequential()

model1.add(Dense(4, input_dim=2, activation='relu'))
model1.add(BatchNormalization())
model1.add(Dense(2,activation='relu'))
model1.add(BatchNormalization())
model1.add(Dense(1,activation='sigmoid'))

model1.summary()
```

- Trainnable params: gama and beta
- non_trainnable param : standard deviation and mean

```python
model1.compile(loss='binary_crossentropy',optimizer='adam',metrics=['accuracy'])
```

```python
history1 = model1.fit(X,y,epochs=200,validation_split=0.2)
```

```python
# plotting loss
plt.plot(history1.history["val_loss"],color="red",label="Val_loss")
plt.plot(history1.history["loss"],color="green",label="loss")
plt.legend()
```

```python
model1.get_weights()
```

- After applying Batch normalization our trainning loss reduces
