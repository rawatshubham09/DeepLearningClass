```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

import tensorflow
from tensorflow import keras
from keras.models import Sequential
from keras.layers import Dense

from sklearn.datasets import make_moons
from sklearn.model_selection import train_test_split
```

```python
%matplotlib inline
```

```python
X,y = make_moons(n_samples=250, noise=0.05, random_state=42)
```

```python
plt.figure(figsize=(8,6))
plt.scatter(X[:,0],X[:,1], c=y, s=100)
```

```python
# Creating Deep Deep Neural Network using Sigmoid activation function

model = Sequential()


model.add(Dense(10, activation="sigmoid", input_dim = 2))
model.add(Dense(10, activation="sigmoid"))
model.add(Dense(10, activation="sigmoid"))
model.add(Dense(10, activation="sigmoid"))
model.add(Dense(10, activation="sigmoid"))
model.add(Dense(10, activation="sigmoid"))
model.add(Dense(10, activation="sigmoid"))
model.add(Dense(10, activation="sigmoid"))
model.add(Dense(10, activation="sigmoid"))
model.add(Dense(10, activation="sigmoid"))
model.add(Dense(10, activation="sigmoid"))
model.add(Dense(10, activation="sigmoid"))
model.add(Dense(1, activation="sigmoid"))
```

```python
model.compile(loss='binary_crossentropy',optimizer='adam',metrics=['accuracy'])
```

```python
# Splitting Trainning and test dataset

X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.20, random_state=42)
```

```python
model.fit(X_train,y_train, epochs=100)
```

```python
old_weights = model.get_weights()[0]
old_weights
```

```python
model.fit(X_train,y_train, epochs=1)
```

```python
new_weights = model.get_weights()[0]
new_weights
```

```python
# Gradient decent values are
gradient_weights = (new_weights - old_weights)/0.01
gradient_weights
```

```python
# Percentage change in new and old weights
percent = abs(100*(new_weights - old_weights)/old_weights)
percent
```

- Gradient weights are very small as it show now improvement in dataset
- slight change in weights smaller then 1% which mean weight are not learning

## Changing Activation function with Relu

```python
model3 = Sequential()


model3.add(Dense(10, activation="relu", input_dim = 2))
model3.add(Dense(10, activation="relu"))
model3.add(Dense(10, activation="relu"))
model3.add(Dense(10, activation="relu"))
model3.add(Dense(10, activation="relu"))
model3.add(Dense(10, activation="relu"))
model3.add(Dense(10, activation="relu"))
model3.add(Dense(10, activation="relu"))
model3.add(Dense(10, activation="relu"))
model3.add(Dense(10, activation="relu"))
model3.add(Dense(10, activation="relu"))
model3.add(Dense(10, activation="relu"))
model3.add(Dense(1, activation="sigmoid"))
```

```python
model3.compile(loss='binary_crossentropy',optimizer='adam',metrics=['accuracy'])
```

```python
old_weights3 = model3.get_weights()[0]
old_weights3
```

```python
model3.fit(X_train,y_train,epochs=100)
```

- Loss is reduce upto 10**-4
- this Show vanishing gradient problem did not occur in the model

```python
new_weights3 = model3.get_weights()[0]
new_weights3
```

```python
# Gradient decent values are
gradient_weights3 = (new_weights3 - old_weights3)/0.01
gradient_weights3
```

```python
# Percentage change in new and old weights
percent3 = abs(100*(new_weights3 - old_weights3)/old_weights3)
percent3
```

- Percentage change varry from 600% inc to 2% inc

### Way to Handel Vanishing Gradient Problem :
    1. By Reduce Model Complexity
    2. Using "Relu" or "leaky_relu" as a activation function
    3. Initilize Proper weights
    4. Batch Normalization
    5. Using Resudal Network like: RESNET in CNN
