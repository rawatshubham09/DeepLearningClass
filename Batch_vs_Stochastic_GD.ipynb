```python
import numpy as np
import pandas as pd
import time
```

```python
# Getting dataset
df = pd.read_csv('Dataset/Social_Network_Ads.csv')
df.head()
```

```python
df = df[['Age','EstimatedSalary','Purchased']]
```

```python
df.head()
```

```python
df.shape
```

```python
X = df.iloc[:,:-1]
y = df.iloc[:,-1]
```

```python
X[:3]
```

```python
from sklearn.preprocessing import StandardScaler
scaler = StandardScaler()

X_scaled = scaler.fit_transform(X)
```

```python
# ANN
import tensorflow
from tensorflow import keras
from keras import Sequential
from keras.layers import Dense
```

```python
model = Sequential()

model.add(Dense(16, activation="relu", input_dim = 2))
model.add(Dense(8, activation="relu"))
model.add(Dense(1, activation="sigmoid"))
```

```python
model.summary()
```

```python
# Gradient decent
model.compile(loss="binary_crossentropy", metrics=["accuracy"])

# starting time
start = time.time()
history = model.fit(X_scaled, y, batch_size=400, epochs=500, validation_split=0.2)
print("Total time taken : ",time.time() - start)
```

```python
# Plotting loss

import matplotlib.pyplot as plt
plt.plot(history.history['loss'])
```

- observations : 
    1. Conversion is smooth but required 500+ epoch.
    2. slow and took lots of computational cost.
    3. took 27 sec to converse below 0.22

```python
# Stochastic decent
model.compile(loss="binary_crossentropy", metrics=["accuracy"])

# starting time
start = time.time()
history = model.fit(X_scaled, y, batch_size=1, epochs=500, validation_split=0.2)
print("Total time taken : ",time.time() - start)
```

```python
# Plotting loss

plt.plot(history.history['loss'])
```

- Observations:
    1. Conversion is not smooth and uncertain.
    2. Required Less no. of Epoch
    3. For same no. of epoch it take more time as no. of loop become (epoch*no_of_rows) but It required less no. of epoch
        to converse model.
    4. It take 7 min and overfitted

```python
# Stochastic decent part 2 with less epoch
model1 = Sequential()

model1.add(Dense(16, activation="relu", input_dim = 2))
model1.add(Dense(8, activation="relu"))
model1.add(Dense(1, activation="sigmoid"))

model1.compile(loss="binary_crossentropy", metrics=["accuracy"])

# starting time
start = time.time()
history = model1.fit(X_scaled, y, batch_size=1, epochs=20, validation_split=0.2)
print("Total time taken : ",time.time() - start)
```

```python
plt.figure(figsize=(10,8))
plt.plot(history.history['loss'])
```

- Observations:
    1. Conversion is not smooth and uncertain.
    2. Required Less no. of Epoch
    3. It take 17 sec

```python
# mini Batch Gradient Descent

# starting time
start = time.time()
history = model1.fit(X_scaled, y, batch_size=40, epochs=20, validation_split=0.2)
print("Total time taken : ",time.time() - start)
```

```python
plt.figure(figsize=(8,6))
plt.plot(history.history['loss'])
```
